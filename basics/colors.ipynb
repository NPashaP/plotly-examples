```python
import plotly.express as px
px.pie(px.data.tips(), 
    values='tip', 
    names='day', 
    color_discrete_map={
        'Thur':'lightcyan',
        'Fri':'cyan',
        'Sat':'royalblue',
        'Sun':'darkblue'
    }
)
```
