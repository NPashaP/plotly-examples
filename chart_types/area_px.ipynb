# Area Chart With Express

```python
import plotly.express as px
px.area(x=[1, 2, 3, 4], y=[0, 2, 3, 5])
```

```python
import plotly.express as px
px.area(px.data.gapminder(), x="year", y="pop", color="continent", line_group="country")
```

```python
import plotly.express as px

px.area(px.data.stocks(indexed=True)-1, facet_col="company", facet_col_wrap=2)
```
