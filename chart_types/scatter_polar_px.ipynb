# Polar Scatter Plot

```python
import plotly.express as px
px.scatter_polar(px.data.wind(), r="frequency", theta="direction")
```

## Color, Size and Symbol

```python
import plotly.express as px
px.scatter_polar(px.data.wind(), 
  r      = "frequency", 
  theta  = "direction", 
  color  = "strength", 
  symbol = "strength",
  size   = "frequency",
  color_discrete_sequence=px.colors.sequential.Plasma_r
)
```

## Start Angle, Range and Direction

```python
import plotly.express as px
px.scatter_polar(r=range(0,90,10), theta=range(0,90,10),
    range_theta=[0,90], start_angle=0, direction="counterclockwise")
```
