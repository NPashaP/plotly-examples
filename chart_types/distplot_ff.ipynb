# Distplot with Figure Factory

## From List Data

```python
import plotly.figure_factory as ff
import numpy as np

# Create distplot with custom bin_size
ff.create_distplot([np.random.randn(200)-2, np.random.randn(200)], ['Group 1', 'Group 2'])
```

## From DataFrame

```python
import plotly.figure_factory as ff
import numpy as np
import pandas as pd

df = pd.DataFrame({'2012': np.random.randn(200),
                   '2013': np.random.randn(200)+1})

ff.create_distplot([df[c] for c in df.columns], df.columns, bin_size=.25)
```

## Bin Size

```python
import plotly.figure_factory as ff
import numpy as np

# Add histogram data
x1 = np.random.randn(200)-2
x2 = np.random.randn(200)
x3 = np.random.randn(200)+2
x4 = np.random.randn(200)+4

group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# Create distplot with custom bin_size
ff.create_distplot([x1, x2, x3, x4], group_labels, bin_size=[.1, .25, .5, 1])
```

## Normal Curve

```python
import plotly.figure_factory as ff
import numpy as np

x1, x2 = np.random.randn(200), np.random.randn(200) + 2

ff.create_distplot([x1, x2], ['Group 1', 'Group 2'],
    curve_type='normal' # override default 'kde'
)
```

## Hide Histogram

```python
import plotly.figure_factory as ff
import numpy as np

x1, x2, x3 = np.random.randn(200) - 1, np.random.randn(200), np.random.randn(200) + 1

ff.create_distplot([x1, x2, x3], ['Group 1', 'Group 2', 'Group 3'], show_hist=False)
```

## Hide Curve

```python
import plotly.figure_factory as ff
import numpy as np

x1, x2, x3 = np.random.randn(200) - 1, np.random.randn(200), np.random.randn(200) + 1

ff.create_distplot([x1, x2, x3], ['Group 1', 'Group 2', 'Group 3'], show_curve=False)
```

## Hide Rug

```python
import plotly.figure_factory as ff
import numpy as np

x1, x2, x3 = np.random.randn(200) - 1, np.random.randn(200), np.random.randn(200) + 1

ff.create_distplot([x1, x2, x3], ['Group 1', 'Group 2', 'Group 3'], show_rug=False)
```

## Custom Color

```python
import plotly.figure_factory as ff
import numpy as np

x1 = np.random.randn(200) - 2
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 2

ff.create_distplot([x1, x2, x3], ['Group 1', 'Group 2', 'Group 3'], 
    colors=['#393E46', '#2BCDC1', '#F66095']
)
```

## Custom

```python
import plotly.figure_factory as ff
import numpy as np

x1, x2 = np.random.randn(26), np.random.randn(26) + .5

rug_text_one = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',
                'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't',
                'u', 'v', 'w', 'x', 'y', 'z']

rug_text_two = ['aa', 'bb', 'cc', 'dd', 'ee', 'ff', 'gg', 'hh', 'ii', 'jj',
                'kk', 'll', 'mm', 'nn', 'oo', 'pp', 'qq', 'rr', 'ss', 'tt',
                'uu', 'vv', 'ww', 'xx', 'yy', 'zz']

# Create distplot with custom bin_size
fig = ff.create_distplot([x1, x2], ['2014', '2015'], 
    bin_size = 0.2,
    rug_text = [rug_text_one, rug_text_two], 
    colors   = ['rgb(0, 0, 100)', 'rgb(0, 200, 200)']
)

fig.update_layout(title_text='Customized Distplot', height=800)
fig.show()
```
