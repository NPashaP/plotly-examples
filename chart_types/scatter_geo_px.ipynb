# Scatter Plots on Maps

```python
import plotly.express as px

px.scatter_geo(px.data.gapminder().query("year == 2007"), 
    locations = "iso_alpha",
    size      = "pop"
)
```

## Categorical Color

```python
import plotly.express as px

px.scatter_geo(px.data.gapminder().query("year == 2007"), 
    locations = "iso_alpha",
    size      = "pop",
    color     = "continent"
)
```

## Projection - Natural Earth

```python
import plotly.express as px

px.scatter_geo(px.data.gapminder().query("year == 2007"), 
    locations = "iso_alpha",
    size      = "pop",
    projection= "natural earth"
)
```

## Tooltip Header (hover_name)

```python
import plotly.express as px

px.scatter_geo(px.data.gapminder().query("year == 2007"), 
    locations = "iso_alpha",
    size      = "pop",
    hover_name="country"
)
```
