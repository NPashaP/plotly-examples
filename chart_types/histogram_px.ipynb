# Histogram With Express

# From DataFrame

```python
import plotly.express as px
px.histogram(px.data.tips(), x="total_bill", y="tip")
```

## Color

```python
import plotly.express as px
px.histogram(px.data.tips(), x="total_bill", color="sex")
```

```python
import plotly.express as px
fig = px.histogram(px.data.tips(), x="total_bill", y="tip")
fig.update_layout(bargap=0.2)
fig
```

## Number of Bins

```python
import plotly.express as px
px.histogram(px.data.tips(), x="total_bill", y="tip", nbins=20)
```

## Percent

```python
import plotly.express as px
px.histogram(px.data.tips(), x="total_bill", y="tip", histnorm='percent')
```

## Probability Density

```python
import plotly.express as px
px.histogram(px.data.tips(), x="total_bill", y="tip", histnorm='probability density')
```

## Average

```python
# Aggregating with avg instead of count
import plotly.express as px
px.histogram(px.data.tips(), x="total_bill", y="tip", histfunc='avg')
```

## Date axis

```python
import plotly.express as px

px.histogram(px.data.stocks(), x="date")
```

## Order

```python
import plotly.express as px
# Here we use a column with categorical data
px.histogram(px.data.tips(), x="day", category_orders={"day":["Thur", "Fri", "Sat", "Sun"]})
```

## Values On Bar

```python
import plotly.express as px
px.histogram(px.data.tips(), x="day", text_auto=True)
```

## Bins From numpy

```python
# Using bins from numpy histogram
import plotly.express as px
import numpy as np

# create the bins
counts, bins = np.histogram(px.data.tips().total_bill, bins=range(0, 60, 5))
bins = 0.5 * (bins[:-1] + bins[1:])

px.bar(x=bins, y=counts, labels={'x':'total_bill', 'y':'count'})
```

## Marginal Rug

```python
import plotly.express as px
px.histogram(px.data.tips(), x="total_bill", y="tip", color="sex", marginal="rug")
```
