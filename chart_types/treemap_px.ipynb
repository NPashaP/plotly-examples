# Treemap With Express

```python
import plotly.express as px
px.treemap(
    names = ["Eve","Cain", "Seth", "Enos", "Noam", "Abel", "Awan", "Enoch", "Azura"],
    parents = ["", "Eve", "Eve", "Seth", "Seth", "Eve", "Eve", "Awan", "Eve"]
)
```

## From Dataframe
If using dataframe for data, use the `path` variable to specify a dummy root and also different columns to use from df.

The size is row count.

```python
import plotly.express as px
px.treemap(px.data.tips(), path=[px.Constant("all"), 'day', 'time', 'sex'])
```

## Specify Aggregate Column

Use `values` paramter to specify column to aggregate.

```python
import plotly.express as px
px.treemap(px.data.tips(), path=[px.Constant("all"), 'day', 'time', 'sex'], values='total_bill')
```

## Discrete Coloring

Using a categorical coloum for `color` results in dicrete coloring

```python
import plotly.express as px
px.treemap(px.data.tips(), path=[px.Constant("all"), 'day', 'time', 'sex'], values='total_bill', color='day')
```

## Continuous Coloring

Using a numeric column for `color` results in coninuous coloring.

```python
import plotly.express as px

px.treemap(px.data.gapminder().query("year == 2007"), path=[px.Constant('world'), 'continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])
```

## Custom Coloring

Use a custom discrete coloring using parameter `color_discrete_map`.

```python
import plotly.express as px
px.treemap(px.data.tips(), path=[px.Constant("all"), 'sex', 'day', 'time'], values='total_bill', color='time',
  color_discrete_map={'(?)':'lightgrey', 'Lunch':'gold', 'Dinner':'darkblue'})
```

## Missing Values in DataFrame

Set missing values in DF to None.

```python
import plotly.express as px
import pandas as pd
vendors = ["A", "B", "C", "D", None, "E", "F", "G", "H", None]
sectors = ["Tech", "Tech", "Finance", "Finance", "Other",
           "Tech", "Tech", "Finance", "Finance", "Other"]
regions = ["North", "North", "North", "North", "North",
           "South", "South", "South", "South", "South"]
sales = [1, 3, 2, 4, 1, 2, 2, 1, 4, 1]
df = pd.DataFrame(
    dict(vendors=vendors, sectors=sectors, regions=regions, sales=sales)
)
df["all"] = "all" # in order to have a single root node
px.treemap(df, path=['all', 'regions', 'sectors', 'vendors'], values='sales')
```
